### Exercise 2: Bar Plot with Colors

```python
# TODO: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']  # Different colors for each bar

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors)

# Customization
plt.title('Bar Plot of Categories and Values', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display the plot
plt.tight_layout()
plt.show()
```
