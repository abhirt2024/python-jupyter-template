### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)
y_line = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
axs[0, 0].plot(x, y_line, label='y = x', color='blue', linewidth=2)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[0, 0].grid(True)

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', s=100)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].grid(True)

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'red', 'orange'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
