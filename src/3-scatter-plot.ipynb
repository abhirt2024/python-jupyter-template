### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', s=100, edgecolor='black')

# Annotation for each point
for i in range(len(x)):
    plt.text(x[i] + 0.1, y[i] + 0.1, f'({x[i]}, {y[i]})', fontsize=12)

# Customization
plt.title('Scatter Plot with Annotations', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
